```python
col_sums = [ d.sum(axis=0)]
sd_sum=pd.DataFrame(col_sums).T
sd_sum=sd_sum.rename(columns={ 0:'sum'})
sd_sum=sd_sum.iloc[0:34,:]
sd_sum.index = map(int, sd_sum.index)
sd_sum.reset_index(inplace = True)
sd_sum=sd_sum.rename(columns={ 'index':'years'})
sd_sum
```

```python
sd_sum.plot(kind="scatter",x='years',y='sum',figsize=(10, 6), color='blue')
```

```python
x = sd_sum['years']      # year on x-axis
y = sd_sum['sum']     # total on y-axis
fit = np.polyfit(x, y, deg=1)
fit
```

```python
sd_sum.plot(kind='scatter', x='years', y='sum', figsize=(10, 6), color='blue')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

# plot line of best fit
plt.plot(x, fit[0] * x + fit[1], color='red') # recall that x is the Years
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# print out the line of best fit
NumofImmigrant='No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1])
NumofImmigrant
```
